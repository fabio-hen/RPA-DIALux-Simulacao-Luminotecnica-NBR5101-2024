```python
import pyautogui
import time
import pyperclip
import subprocess
import pandas as pd
import keyboard
from unidecode import unidecode
from pathlib import Path
import logging

logging.basicConfig(level=logging.INFO)

def log_info(mensagem):
    logging.info(mensagem)



pyautogui.FAILSAFE = True

df = pd.read_excel(r'C:\Users\engen\OneDrive\KMR-Operacional\Treinamentos e procedimentos\Automação\Logradouro.xlsx')
display(df)

caminho_pasta = Path(r"C:\Users\engen\OneDrive\Área de Trabalho\Simulações")

def Classesilum(Clas_via, Clas_ped):
    seguranca = 1.00
    dic_ClassVia = {"C0": [50.00, 0.38], 
                    "C1": [30.00, 0.38], 
                    "C2": [20.00, 0.28], 
                    "C3": [15.00, 0.18],
                    "C4": [10.00, 0.18], 
                    "C5": [7.50,  0.18]}

    iluminvia = dic_ClassVia[Clas_via][0] * seguranca
    fatvia = dic_ClassVia[Clas_via][1] * seguranca

    dic_ClassPed = {"P1": [20.00, 4.0], 
                    "P2": [15.00, 3.0], 
                    "P3": [10.00, 2.0],
                    "P4": [7.50,  1.5],
                    "P5": [5.00,  1.0], 
                    "P6": [3.00,  0.6]}


    iluminped = dic_ClassPed[Clas_ped][0] * seguranca
    fatped = dic_ClassPed[Clas_ped][1] * seguranca
    return iluminvia, fatvia, iluminped, fatped

# GUARDA AS INFORMAÇÕES DA VIA EM VARIAVEIS
def infos(tabela, linha):
    Num = tabela['N°'][linha]
    Via = tabela['VIA'][linha]
    Clas_via = tabela['Classificação Via'][linha]
    Clas_ped = tabela['Classificação Pedestre'][linha]
    Pav = tabela['Pavimento'][linha]                      
    Caladj  = round(float(tabela['Calçada adjacente (m)'][linha]), 2)               
    Numfaixasviaadj = round(int(tabela['Pista 01  Número de faixas via adj.'][linha]))
    Pista1 = round(float(tabela['Pista 01   largura total (m)'][linha]), 2)     
    CantCentral = round(float(tabela['Largura canterio central'][linha]), 2)             
    Numfaixasviaop = round(int(tabela['Pista 02  Número de faixas via oposta'][linha]))  
    Pista2 = round(float(tabela['Pista 02   largura total (m)'][linha]), 2)      
    Calop = round(float(tabela['Calçada oposta (m)'][linha]), 2)                
    DispPostes = tabela['Disposição dos postes'][linha]                   
    Vao = round(float(tabela['Vão entre postes (m)'][linha]), 2)
    Recuo = round(float(tabela['Recuo do poste a via (m)'][linha]), 2)
    Altura = round(float(tabela['Altura de Ponto de luz'][linha]), 2)
    Proje = round(float(tabela['Projeção'][linha]), 2)
    Pendor = round(float(tabela['PENDOR'][linha]), 2)
    LumPorPoste = tabela['Luminárias por poste'][linha]
    return Num, Via, Clas_via, Clas_ped, Pav, Caladj, Numfaixasviaadj, Pista1, CantCentral, Numfaixasviaop, Pista2, Calop, DispPostes, Vao, Recuo, Altura, Proje, Pendor, LumPorPoste 


def encontrar_imagem(imagem, tentativas=5):
    for _ in range(tentativas):
        encontrou = pyautogui.locateOnScreen(imagem, grayscale=True, confidence=0.9)
        if encontrou:
            return encontrou
        time.sleep(0.5)
    raise Exception(f'Imagem {imagem} não encontrada após {tentativas} tentativas')


def escrever_texto(texto):
    pyautogui.write(str(texto))

def colar_texto(texto):
    pyperclip.copy(texto)
    pyautogui.hotkey("ctrl", "v")
    
def direita(posicoes_imagem):
    return posicoes_imagem[0] + posicoes_imagem[2], posicoes_imagem[1] + posicoes_imagem[3]/2

def embaixo(posicoes_imagem):
    return posicoes_imagem[0] + posicoes_imagem[2]/2, posicoes_imagem[1] + posicoes_imagem[3]

def up(n):
    for i in range(n):
        pyautogui.hotkey('Up')

def enter(n):
    for i in range(n):
        pyautogui.hotkey('Enter')

def down(n):
    for i in range(n):
        pyautogui.hotkey('down')

def tab(n):
    for i in range(n):
        pyautogui.hotkey('Tab')
       
# ABRE O DIALUX
def abrir_dialux():
    subprocess.Popen([r"C:\Program Files (x86)\DIALux\DIALux.exe"])
    #esperar carregar o dialux
    while not pyautogui.locateOnScreen('dialux_aberto.png'):
        time.sleep(3)

# ADICIONA A VIA 1
def adc_via1():
    encontrou = encontrar_imagem("editar.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("largura.png")
    pyautogui.click(pyautogui.center(encontrou))
    pyautogui.hotkey('ctrl', 'a')
    escrever_texto((Pista1))
    tab(1)
    escrever_texto((Numfaixasviaadj))
    tab(2)
    pyautogui.hotkey('space')

    if Pav == 'Asfalto':
        enter(1)

    if Pav == 'Piso Intertravado':
        up(2)
        enter(1)
               	
    if Pav == 'Chão Batido':
        down(20)
        enter(1)
    encontrou = encontrar_imagem("campoavali.png")
    pyautogui.click(pyautogui.center(encontrou))

    tab(1)
    pyautogui.hotkey('space')
    pyautogui.write("CE")
    if Clas_via == 'C0' or Clas_via == 'C1':
        enter(1)
    if Clas_via == 'C2':
        down(1)
        enter(1)
    if Clas_via == 'C3':
        down(2)
        enter(1)
    if Clas_via == 'C4':
        down(3)
        enter(1)
    if Clas_via == 'C5':
        down(4)
        enter(1) 

    encontrou = encontrar_imagem("pontos.png")
    pyautogui.click(pyautogui.center(encontrou))   

    tab(1)
    escrever_texto("17")
    tab(1)
    escrever_texto(Numfaixasviaadj * 5)
    enter(1)

def adc_via2():
    encontrou = encontrar_imagem("editar.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("pistaderodagem.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("largura.png")
    pyautogui.click(pyautogui.center(encontrou))

    pyautogui.hotkey('ctrl', 'a')
    escrever_texto((Pista2))
    tab(1)
    escrever_texto((Numfaixasviaop))
    tab(2)
    pyautogui.hotkey('space')

    if Pav == 'Asfalto':
        enter(1)

    if Pav == 'Piso Intertravado':
        up(2)
        enter(1)

    if Pav == 'Chão Batido':
        down(20)
        enter(1)
    encontrou = encontrar_imagem("campoavali.png")
    pyautogui.click(pyautogui.center(encontrou))

    tab(1)
    pyautogui.hotkey('space')
    pyautogui.write("CE")
    if Clas_via == 'C0' or Clas_via == 'C1':
        enter(1)
    if Clas_via == 'C2':
        down(1)
        enter(1)
    if Clas_via == 'C3':
        down(2)
        enter(1)
    if Clas_via == 'C4':
        down(3)
        enter(1)
    if Clas_via == 'C5':
        down(4)
        enter(1)

    encontrou = encontrar_imagem("pontos.png")
    pyautogui.click(pyautogui.center(encontrou))   

    tab(1)
    escrever_texto("17")
    tab(1)
    escrever_texto(Numfaixasviaadj * 5)
    enter(1)
    

def adc_canteiro():
    encontrou = encontrar_imagem("editar.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("canteiro.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("larguracanteiro.png")
    pyautogui.click(pyautogui.center(encontrou))
    pyautogui.hotkey('ctrl', 'a')
    escrever_texto((CantCentral))
    tab(2)
    escrever_texto(0.15)
    enter(1)

def adc_cal1():
    encontrou = encontrar_imagem("editar.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("passeio.png")
    pyautogui.click(pyautogui.center(encontrou))

    # clicar no meio da imagem
    encontrou = encontrar_imagem("movebaixo.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("largurapasseio.png")
    pyautogui.click(pyautogui.center(encontrou))
    pyautogui.hotkey('ctrl', 'a')
    escrever_texto(Caladj)
    tab(2)
    escrever_texto(0.15)
    encontrou = encontrar_imagem("campoavali.png")
    pyautogui.click(pyautogui.center(encontrou))
    tab(1)
    pyautogui.hotkey('space')
    pyautogui.write("CE")
    
    if Clas_ped == 'P1':
        enter(1)
    if Clas_ped == 'P2':
        enter(1)
    if Clas_ped == 'P3':
        down(1)
        enter(1)
    if Clas_ped == 'P4':
        down(2)
        enter(1)
    if Clas_ped == 'P5':
        down(3)
        enter(1)
    if Clas_ped == 'P6':
        down(3)
        enter(1)
    # clicar no meio da imagem
    encontrou = encontrar_imagem("pontos.png")
    pyautogui.click(pyautogui.center(encontrou))


    tab(1)
    escrever_texto("17")
    tab(1)
    if Caladj < 3:
        faixa = 1
    else:
        faixa = 2
    escrever_texto((faixa))
    enter(1)


def adc_cal2():
    encontrou = encontrar_imagem("editar.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("passeio.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("largurapasseio.png")
    pyautogui.click(pyautogui.center(encontrou))
    pyautogui.hotkey('ctrl', 'a')
    escrever_texto(Calop)
    tab(2)
    escrever_texto(0.15)
    encontrou = encontrar_imagem("campoavali.png")
    pyautogui.click(pyautogui.center(encontrou))

    tab(1)
    pyautogui.hotkey('space')
    pyautogui.write("CE")
    
    if Clas_ped == 'P1':
        enter(1)
    if Clas_ped == 'P2':
        enter(1)
    if Clas_ped == 'P3':
        down(1)
        enter(1)
    if Clas_ped == 'P4':
        down(2)
        enter(1)
    if Clas_ped == 'P5':
        down(3)
        enter(1)
    if Clas_ped == 'P6':
        down(3)
        enter(1)

    # clicar no meio da imagem
    encontrou = encontrar_imagem("pontos.png")
    pyautogui.click(pyautogui.center(encontrou))

    tab(1)
    escrever_texto("17")
    tab(1)
    if Caladj < 3:
        faixa = 1
    else:
        faixa = 2
    escrever_texto((faixa))
    enter(1)

def adc_parametros():
    pyperclip.copy(Via)
    # clicar no meio da imagem
    encontrou = encontrar_imagem("nomerua.png")
    pyautogui.click(pyautogui.center(encontrou))
    via_sem_acento = unidecode(Via)
    pyautogui.write(via_sem_acento)
    enter(1)
    tab(2)
    escrever_texto(0.7)
    enter(1)

    #SELEÇÃO DAS LUMINARIAS
    tab(2)
    pyautogui.hotkey('ctrl', 'a')
    tab(1), enter(1)
    tab(3), enter(1)
    pyautogui.click(x=1400, y=500)

    #SELEÇÃO DA DISPOÇÃO DOS POSTES
    tab(2), pyautogui.hotkey('space')
    if DispPostes == 'Unilateral':
        enter(1)
    if DispPostes == 'Bilateral Frente a Frente':
        encontrou = encontrar_imagem("bifrentefrente.png")
        pyautogui.click(pyautogui.center(encontrou))
    if DispPostes == 'Bilateral Alternado':
        encontrou = encontrar_imagem("bialternado.png")
        pyautogui.click(pyautogui.center(encontrou))
    if DispPostes == 'Canteiro Central':
        encontrou = encontrar_imagem("canteirocentral.png")
        pyautogui.click(pyautogui.center(encontrou))
    pyautogui.click(x=1400, y=500)

    #SELEÇÃO DO VÃO ENTRE POSTES
    tab(2)
    pyautogui.hotkey('space')
    tab(1)
    escrever_texto(Vao)
    pyautogui.click(x=1400, y=500)

    #SELEÇÃO DA ALTURA
    tab(3)
    escrever_texto(Altura)
    pyautogui.click(x=1400, y=500)

    #SELEÇÃO DA INCLICAÇÃO
    tab(2), pyautogui.hotkey('space')
    pyautogui.click(x=1400, y=500)

    #SELEÇÃO DO PENDOR
    tab(3)
    escrever_texto(Pendor)
    pyautogui.click(x=1400, y=500)

    #SELEÇÃO DO RECUO
    tab(1)
    escrever_texto(Recuo)
    pyautogui.click(x=1400, y=500)

    tab(3)

    #NORMAS PASSEIO 2
    if Calop > 0:
        escrever_texto(iluminped)
        pyautogui.click(x=1400, y=500)
        tab(1), enter(1), pyautogui.hotkey('space')
        tab(1)
        escrever_texto(fatped)
        pyautogui.click(x=1400, y=500)
        pyautogui.hotkey('tab'), enter(1), pyautogui.hotkey('space')
        if  Clas_ped == 'P1' or Clas_ped == 'P2' or  Clas_ped == 'P3':
            tab(1), enter(1), pyautogui.hotkey('space')
        pyautogui.click(x=1400, y=500)
        tab(2)

    #NORMAS VIA 2
    if Pista2 > 0:
        escrever_texto(iluminvia)
        pyautogui.click(x=1400, y=500)
        tab(1)
        escrever_texto(fatvia)
        pyautogui.click(x=1400, y=500)
        tab(1), enter(1), pyautogui.hotkey('space')

        if Clas_via == 'C0' or  Clas_via == 'C1' or Clas_via == 'C2' or Clas_via == 'C3':
            tab(1), enter(1), pyautogui.hotkey('space')
        pyautogui.click(x=1400, y=500)
        tab(2)

    #NORMAS VIA 1
    if Pista1 > 0:
        escrever_texto(iluminvia)
        pyautogui.click(x=1400, y=500)
        tab(1)
        escrever_texto(fatvia)
        pyautogui.click(x=1400, y=500)
        tab(1), enter(1), pyautogui.hotkey('space')
        if Clas_via == 'C0' or  Clas_via == 'C1' or Clas_via == 'C2' or Clas_via == 'C3':
            tab(1), enter(1), pyautogui.hotkey('space')
        pyautogui.click(x=1400, y=500)
        tab(2)

    #NORMAS PASSEIO 1
    if Caladj > 0:  
        escrever_texto(iluminped)
        pyautogui.click(x=1400, y=500)
        tab(1), enter(1), pyautogui.hotkey('space')
        tab(1)
        escrever_texto(fatped)
        pyautogui.click(x=1400, y=500)
        tab(1), enter(1), pyautogui.hotkey('space')
        if  Clas_ped == 'P1' or Clas_ped == 'P2' or  Clas_ped == 'P3':
            tab(1), enter(1), pyautogui.hotkey('space')
        pyautogui.click(x=1400, y=500)


#Guardar
def salvar():  
    pyautogui.hotkey('ctrl', 's')
    nome = rf'{caminho_pasta}\{Num} - {Via}.dlx'
    colar_texto(nome)
    enter(1)

def relatorio():
    encontrou = encontrar_imagem("edicao.png")
    pyautogui.click(pyautogui.center(encontrou))  
    encontrou = encontrar_imagem("maisedicoes.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("kmr.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("visualizar.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("verpagina.png")
    encontrou = encontrar_imagem("okverpagina.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("visuaberto.png")
    encontrou = encontrar_imagem("fecharvisu.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("gerarrela.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("tudo.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("okgerarrela.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("salvarcomo.png")
    nome = rf'{caminho_pasta}\{Num} - {Via}.pdf'
    colar_texto(nome)
    enter(1)
    encontrou = encontrar_imagem("okpdf.png")
    time.sleep(3)
    enter(1)



def novo():
    pyautogui.hotkey('ctrl', 'n')
    while not pyautogui.locateOnScreen('editorrua.png'):
        time.sleep(0.5)   

def optimizar():
    pyautogui.click(x=1400, y=500)
    pyautogui.scroll(-1200)
    encontrou = encontrar_imagem("inioptimizar.png")
    pyautogui.click(pyautogui.center(encontrou))
    encontrou = encontrar_imagem("outrosresul.png")
    if pyautogui.locateOnScreen("xfalha.png", grayscale=True, confidence=0.9):
        pass
    else:
        encontrou = encontrar_imagem("outrosresul.png")
        pyautogui.click(pyautogui.center(encontrou)) 
        encontrou = encontrar_imagem("maxoutrosres.png")
        pyautogui.click(pyautogui.center(encontrou)) 
        pyautogui.click(x=1780,y=963)
        encontrou = encontrar_imagem("potencia.png")
        pyautogui.click(pyautogui.center(encontrou)) 
        encontrou = encontrar_imagem("pripotencia.png")
        pyautogui.click(embaixo(encontrou))
        encontrou = encontrar_imagem("aplicar.png")
        pyautogui.click(pyautogui.center(encontrou)) 
        encontrou = encontrar_imagem("criarrua.png")
        pyautogui.click(pyautogui.center(encontrou))
        tab(4)
        pyautogui.hotkey('space')
        tab(2)
        pyautogui.hotkey('space')
        time.sleep(3)
        while  pyautogui.locateOnScreen("calculando.png", grayscale=True, confidence=0.9):  # reconhecimento de imagem
            time.sleep(0.2)
        encontrou = encontrar_imagem("minimilumi.png")
        pyautogui.click(pyautogui.center(encontrou))
        encontrou = encontrar_imagem("ruacriada.png")
        pyautogui.click(pyautogui.center(encontrou))
        encontrou = encontrar_imagem("nomerua3.png")
        pyautogui.click(direita(encontrou))
        pyautogui.hotkey('ctrl', 'a')
        colar_texto(Via)
        relatorio()


log_info("Iniciando automação")
pyautogui.alert('O código vai começar. Não mexa em NADA enquanto o código tiver rodando. Quando finalizar, eu te aviso')
pyautogui.PAUSE = 0.2
abrir_dialux() 

while True:
    for linha in range(0, len(df)):
                          
        Num, Via, Clas_via, Clas_ped, Pav, Caladj, Numfaixasviaadj, Pista1, CantCentral, Numfaixasviaop, Pista2, Calop, DispPostes, Vao, Recuo, Altura, Proje, Pendor, LumPorPoste = infos(df, linha)
        iluminvia, fatvia, iluminped, fatped = Classesilum (Clas_via, Clas_ped)

        #encontrou = encontrar_imagem("editorrua.png")
        #pyautogui.click(pyautogui.center(encontrou))
        encontrou = encontrar_imagem("proje.png")
        pyautogui.rightClick(pyautogui.center(encontrou))
        tab(5)
        enter(1)
        encontrou = encontrar_imagem("pro1.png")
        pyautogui.click(direita(encontrou))
        pyperclip.copy(Via)
        pyautogui.hotkey('ctrl', 'a')
        pyautogui.hotkey('ctrl', 'v')
        tab(1)
        Descri = f'''Referência Normativa: NBR 5101:2018\n
        
        Classificação da via: {Clas_via}\n
        Classificação dos Passeios: {Clas_ped}'''
        colar_texto(Descri)

        if Pista1 > 0:
            adc_via1()

        if Pista2 > 0:
            adc_via2()

        if CantCentral > 0:
            adc_canteiro()

        if Caladj > 0:
            adc_cal1()

        if Calop > 0:
            adc_cal2()

        adc_parametros()
        optimizar()
        salvar()
        novo()
        #no segundo projeto ele para
        #colocar passoas para gera relatorio no padrao
    
    pyautogui.alert('O código terminou, pode pegar o seu computador de volta')
```
